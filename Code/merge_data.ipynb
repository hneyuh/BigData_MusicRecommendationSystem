```python
import pandas as pd
import os
```

```python
path = 'D:\BIGDATA_FinalProject/'
```

```python
frames = []
for file in os.listdir(path):
    filepath = path + file
    df1 = pd.read_csv(filepath)
    frames.append(df1) 
    result = pd.concat(frames)
df = result
```

```python
df.drop(columns=['Unnamed: 0'], inplace= True)
```

```python
df.head()
```

```python
df.drop(columns=['Unnamed: 0.1'], inplace= True)
```

```python
df
```

```python
df['title']=df['title'].apply(str)
```

```python
df.info()
```

```python
gen1 = []
for i in df.genres:
    string1 = i.replace("'", "").replace(" ","_").replace(",_", ", ")
    gen1.append(string1)
```

```python
df['genres_list'] = gen1
```

```python
gen = []
for i in df.genres:
    string = i.replace('[', '').replace(']', '').replace("'", "").replace(" ","_").replace(",_", ", ").replace(", ", " ")
    gen.append(string)
```

```python
df['genres'] = gen
```
